```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Sig.csv')
df
```

```python
df = df.replace('--', np.nan)
df
```

```python
df = df.transpose()
df
```

```python
x_lst = []
y_lst = []
num_set = int(df.shape[0]/2)
for i in range(num_set):   
    x = df.iloc[2*i].values.astype(np.float32)
    x = x[~np.isnan(x)] # ~: bitwise not 取反運算子(True變False)
    y = df.iloc[2*i+1].values.astype(np.float32)
    y = y[~np.isnan(y)]
    x_lst.append(x)
    y_lst.append(y)
```

```python
for i in range(5):
    plt.figure()
    plt.plot(x_lst[i], y_lst[i], '-o')
    plt.show()
```

```python
def sig5(x):
    A1 = -0.0113177013264537
    A2 = 0.951482622149164
    x0 = -0.172518688334704
    dx = 0.531396293380814
    return (A1-A2) / (1+np.exp((x-x0)/dx)) + A2 

def sig(x):
    return 1 / (1 + np.exp(-x))

X = np.linspace(-4, 4, 100)
X_sig5 = sig5(X)
X_sig = sig(X)
plt.figure()
plt.plot(X, X_sig5, 'r', label='Fitted sigmoid function')
plt.plot(X, X_sig, 'k', label='Ideal sigmoid function')
plt.legend()
plt.show()
```
