```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('SOT_1.csv')
print(df.shape)
df.head(10)
```

```python
df = df.replace('--', np.nan)
df.tail(5)
```

```python
df = df.transpose()
df.head(5)
```

```python
x_lst = []
y_lst = []
num_set = int(df.shape[0]/2)
for i in range(num_set):   
    x = df.iloc[2*i].values.astype(np.float32)
    x = x[~np.isnan(x)] # ~: bitwise not 取反運算子(True變False)
    y = df.iloc[2*i+1].values.astype(np.float32)
    y = y[~np.isnan(y)]
    x_lst.append(x)
    y_lst.append(y)
```

```python
# cmap = plt.get_cmap('viridis')
for i in range(28):
    plt.figure()
    plt.plot(x_lst[i], y_lst[i], '-o', color=np.random.rand(3))
    plt.show()
```
